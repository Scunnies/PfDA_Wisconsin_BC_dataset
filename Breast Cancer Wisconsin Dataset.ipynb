# 3. The Wisconsin Breast Cancer Dataset
***

Breast cancer is one of the most common diseases in women aged 40 to 59 years. Second, only to lung cancer, breast cancer is the most common cancer globally in terms of incidence. Breast cancer is estimated to affect one in eight women at some point in their lives. Men can also be diagnosed with breast cancer but women are 100 times more likely to be affected and one woman in 44000 dies from breast cancer each year.

The best preventive method against breast cancer is early detection and when detected early, the patient has a 96% chance of survival before it spreads.  The only one way to know for sure whether a mass in the breast is benign or malignant is to take a sample by biopsy. 

The Wisconsin Breast Cancer dataset was obtained by imaging a needle-tip-wide breast mass with biosip by Dr. William Wolberg, an employee of Wisconsin Hospital, and digitizing these images by William Nick Street, one of the researchers of the University of Wisconsin Computer Sciences Department, in November 1995.

# Cleaning and Preparing the Dataset

First I will check my dataset for null values. 

```python
df.isnull().sum()
```

The only column with null values is the last one, 'Unnamed: 32' so I will drop this along with the ID column which isn't relevant for the purposes of this assignment and then check that I now have 31 columns instead of 33. 

```python
df.drop('id',axis=1,inplace=True)
df.drop('Unnamed: 32',axis=1,inplace=True)
df.head(6)
```

I'd prefer to have diagnosis as a numeric value rather than categorical so that Malignant is represented by 1 and benign by O.

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```
